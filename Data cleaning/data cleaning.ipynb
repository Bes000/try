```python
import pandas as pd
import numpy as np
import xlrd

years = list(map(int, np.linspace(1960, 2022, int(2023 - 1960))))
years
```

```python

gdp_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NY.GDP.MKTP.CD_DS2_en_excel_v2_93.xls")
gdp_data = gdp_data[gdp_data['Data Source'] == 'Saudi Arabia']
gdp_data = gdp_data.iloc[:,4:]
gdp_data.rename(columns= {i : j for i, j in zip(gdp_data.columns, years)}, inplace= True )
gdp_data.rename(index= {208: 'GDP'}, inplace= True)
gdp_data = gdp_data.transpose()
gdp_data
```

```python
exp_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NE.EXP.GNFS.CD_DS2_en_excel_v2_383.xls")
exp_data = exp_data[exp_data['Data Source'] == 'Saudi Arabia']
exp_data = exp_data.iloc[:,4:]
exp_data.rename(columns= {i : j for i, j in zip(exp_data.columns, years)}, inplace= True )
exp_data.rename(index= {208 : 'EXP'}, inplace= True)
exp_data.drop(columns= 'Unnamed: 67', inplace= True)
exp_data = exp_data.transpose()

exp_data

```

```python
con_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NE.CON.TOTL.CD_DS2_en_excel_v2_532.xls")
con_data = con_data[con_data['Data Source'] == 'Saudi Arabia']
con_data = con_data.iloc[:,4:]
con_data.rename(columns= {i : j for i, j in zip(con_data.columns, years)}, inplace= True )
con_data.rename(index= {208 : 'CON'}, inplace= True)
con_data.drop(columns= 'Unnamed: 67', inplace= True)
con_data = con_data.transpose()
con_data
```

```python
gov_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NE.CON.GOVT.CD_DS2_en_excel_v2_42485.xls")
gov_data = gov_data[gov_data['Data Source'] == 'Saudi Arabia']
gov_data = gov_data.iloc[:,4:]
gov_data.rename(columns= {i : j for i, j in zip(gov_data.columns, years)}, inplace= True )
gov_data.rename(index= {208 : 'GOV_EXP'}, inplace= True)
gov_data.drop(columns= 'Unnamed: 67', inplace= True)
gov_data = gov_data.transpose()
gov_data
```

```python

inv_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NE.GDI.TOTL.CD_DS2_en_excel_v2_44593.xls")
inv_data = inv_data[inv_data['Data Source'] == 'Saudi Arabia']
inv_data = inv_data.iloc[:,4:]
inv_data.rename(columns= {i : j for i, j in zip(inv_data.columns, years)}, inplace= True )
inv_data.rename(index= {208 : 'INV'}, inplace= True)
inv_data.drop(columns= 'Unnamed: 67', inplace= True)
inv_data = inv_data.transpose()
inv_data

```

```python
imp_data = pd.read_excel(r"C:\Users\SoftLaptop\Desktop\econometrics\GDP\API_NE.IMP.GNFS.CD_DS2_en_excel_v2_43867.xls")


imp_data = imp_data[imp_data['Data Source'] == 'Saudi Arabia']
imp_data = imp_data.iloc[:,4:]
imp_data.rename(columns= {i : j for i, j in zip(imp_data.columns, years)}, inplace= True )
imp_data.rename(index= {208 : 'IMP'}, inplace= True)
imp_data.drop(columns= 'Unnamed: 67', inplace= True)
imp_data = imp_data.transpose()
imp_data
```

```python

net_exp = pd.DataFrame(index = years, columns=['Net_exp'])
net_exp['Net_exp'] = exp_data['EXP']-imp_data['IMP']
net_exp
```

```python
df = pd.DataFrame(columns= ['GDP', "GOV_EXP", 'CON', 'INV'], index= years)
df.GDP =  gdp_data['GDP']

df['GOV_EXP'] = gov_data['GOV_EXP']
df['CON'] = con_data['CON']
df['INV'] = inv_data['INV']
df['NET_EXP'] = net_exp['Net_exp']
df['EXP'] = exp_data['EXP']
df['IMP'] = imp_data['IMP']
df
```

```python
df.to_excel('saudi.xlsx')
```
